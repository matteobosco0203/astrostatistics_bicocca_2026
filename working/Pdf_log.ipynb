```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.random.uniform(0.1,10,10000)
plt.hist(x, density=True, histtype="step")
```

```python
y=np.log10(x)
plt.hist(y, density=True, histtype="step")
a=np.linspace(-1,1,100)
plt.plot(a,10**a*np.log(10)/(10-0.1))
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
median_x=np.median(x)
median_y=np.median(y)

print(np.log10(mean_x),mean_y)
print(np.log10(median_x),median_y)
```
