# Nasty Integral

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
```

```python
N=10000
sigma=2
mu=0

distG = scipy.stats.norm(loc=mu, scale=sigma)
x = np.abs(distG.rvs(N))
```

```python
def f(x, sigma):
    y=x**3*sigma*0.5*np.sqrt(2*np.pi)
    return y

def Monte(x,sigma,f,N):
    inte=np.sum(f(x,sigma))/N
    return inte
```

```python
inte=Monte(x,sigma,f,N)
print(inte)
```

# Kicking Horses

```python
x=np.array([0,1,2,3,4])
y=np.array([109,65,22,3,1])
y=y/np.sum(y)
```

```python

plt.scatter(x,y)
mu=np.sum(y)/len(y)
sigma=np.sqrt(1/(len(y)-1)*np.sum((y-mu)**2))
for mu in [0.1,0.5,1,2]:
    p = scipy.stats.poisson(mu).pmf(x)
    plt.plot(x, p,label='$\mu='+str(mu)+"$")
plt.legend()
```

```python
goodmu = np.average(x,weights=y)
plt.scatter(x, y)

p = scipy.stats.poisson(goodmu).pmf(x)
plt.plot(x, p)
```
