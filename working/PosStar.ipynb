```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu=1
sigma=0.2
N=50
data=norm.rvs(mu, sigma, size=N)
print(data)
MLE=np.mean(data)
```

```python
gauss_array=np.array([0])
x=np.linspace(0,2,100)
g=1
for i in range(N):
    dist=norm(data[i],sigma)
    y=dist.pdf(x)
    plt.plot(x,y)
    g=g*y

```

```python
plt.plot(x,g/1000)
```

```python
idx=np.argmax(g)
mu=x[idx]
print(mu, MLE)
err=np.std(data)/np.sqrt(N)
print(err)
```

```python
logg=np.log(g)
g2=np.diff(logg,2)
print(g2)
x2=np.delete(x,[0,1])
g3=g2/np.abs(mu-MLE)
plt.plot(x2,g2/np.abs(mu-MLE))
```

```python
print(1/np.sqrt(-g3))
print(err)
sigma_new=np.sqrt(-g3)[0]
dist=norm(mu,1/sigma_new) #1/ ?? idk looked like it from the theory
print(dist)
y2=dist.pdf(x2)
print(y2)
plt.plot(x2,y2)

print(sigma/np.sqrt(N))
```
